```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Enable inline plotting
%matplotlib inline

def projectile_motion(v0, theta, b, m, g, dt, t_final):

    # Initial conditions
    vx = v0 * np.cos(theta)
    vy = v0 * np.sin(theta)
    x = 0
    y = 0

    # Arrays to store results
    t = np.arange(0, t_final, dt)
    x_arr = [x]
    y_arr = [y]

    # Simulation loop using Euler's Method
    for _ in t[1:]:  # Start from 1 to skip initial condition
        v_mag = np.sqrt(vx**2 + vy**2)  # magnitude of the velocity
        ax = -(b/m) * v_mag * vx        # x-direction
        ay = -g - (b/m) * v_mag * vy    # y-direction

        # Update velocity
        vx = vx + ax * dt
        vy = vy + ay * dt

        # Update position
        x = x + vx * dt
        y = y + vy * dt

        x_arr.append(x)
        y_arr.append(y)

    return t, np.array(x_arr), np.array(y_arr)
```

```python
# Example Usage and Plotting
v0 = 30  # m/s
theta = np.pi/4  # 45 degrees in radians
b = 0.1  # air resistance coefficient
m = 1  # kg
g = 9.81  # m/s^2
dt = 0.01  # seconds
t_final = 5  # seconds

# Simulate with air resistance
t, x, y = projectile_motion(v0, theta, b, m, g, dt, t_final)
```

```python
# Plotting projectile motion with air resistance
plt.plot(x, y)
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Projectile Motion with Air Resistance")
plt.grid(True)
plt.show()  # Display the plot in Colab
```

```python
# Example with no air resistance for comparison
b = 0  # air resistance coefficient
t_no_air, x_no_air, y_no_air = projectile_motion(v0, theta, b, m, g, dt, t_final)

# Plotting projectile motion without air resistance
plt.plot(x_no_air, y_no_air)
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Projectile Motion without Air Resistance")
plt.grid(True)
plt.show()  # Display the plot in Colab
```
